### **Prepare the data according to the Apriori Algorithm Expectation**

```python
Retail_Pivot_Data = pd.pivot_table(data = Retail_Data, values='Quantity', index='InvoiceNo', columns='Description').fillna(0)
Retail_Pivot_Data
```

```python
Retail_Pivot_Data = Retail_Pivot_Data.applymap(lambda x: 1 if x >0 else 0)
Retail_Pivot_Data 
```

## **Model Building**

```python
frequent_item_list = apriori(df = Retail_Pivot_Data , min_support=0.03, use_colnames=True, max_len=None, n_jobs=1)
frequent_item_list
```

```python
Ass_Rules = association_rules(df = frequent_item_list , metric="confidence", min_threshold=0.05)
Ass_Rules
```

##**Observation**
When min_support is 3%

GREEN REGENCY TEACUP AND SAUCER and ROSES REGENCY TEACUP AND SAUCER are best associates with 74% confidence
